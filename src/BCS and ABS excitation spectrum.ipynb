```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#facciamo l excitation spectruum
x = np.linspace(0,3, 100, endpoint=True)
def exc(x):
    cos = [(np.cos(obj * np.pi) + 4) for obj in x if obj < 1]
    quad = [(obj -1)**2 + 3 for obj in x if obj >= 1]
    return np.array(cos + quad)
```

```python
#facciamo gli abs
ABS = []
for i in range(5):
    level = (0.5*i + 0.5) * np.ones((1, len(x)), dtype=float)
    ABS.append(level)
ABS = np.array(ABS)
ABS.shape
```

```python
#plottiamo
xticks = []
for obj in x:
    if obj == 1:
        xticks.append(f"$k_F$")
    else:
        xticks.append(None)

plt.figure(figsize = (9,7), dpi = 300)
plt.plot(x, exc(x), lw = 3, color = "orange", label = "excitation spectrum")
for i in range(5):
    lab = "bound states"
    boool = not(i or i)
    plt.plot(x, ABS[i,0,:], lw = 3, color = "#e67300", label = lab*boool, ls = 'solid')
plt.yticks([])
plt.xticks(x, xticks, fontsize = 15)
plt.ylabel(f"$\\epsilon$(k)", fontsize = 15)
plt.xlabel("k", fontsize = 15)
plt.title("Energy spectrum: Discrete levels and Continuum", fontweight='bold')
plt.gca().set_facecolor('#e7e4e4')
plt.legend()
plt.savefig("Energy_spectrum_Discrete_levels_and_Continuum.png", dpi=300, bbox_inches='tight')
plt.show()
```
